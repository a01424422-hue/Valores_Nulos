```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name = "2020")
data.head(5)
```

```python
data2= pd.read_excel('Gastos y costos 20-23.xlsx', skiprows=5)
data2
```

```python
#Revisar columnas con datos nulos
data2.isnull().sum()
```

```python
#Valor en concreto - folio
data2["FOLIO"]= data2["FOLIO"].fillna(121212)
data2
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Valor en concreto - gasto
data2["GASTO"]= data2["GASTO"].fillna("SIN GASTO")
data2
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Valor en concreto - tarjeta de credito
data2["TC"]= data2["TC"].fillna(0)
data2
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Media - importe
data2["IMPORTE"]=data2["IMPORTE"].fillna(round(data2["IMPORTE"].mean(),1))
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Importe x .16 - IVA
data2["IVA"]= data2["IVA"].fillna(data2["IMPORTE"] * .16)
data2
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Valor en concreto - tipo
data2["TIPO"]= data2["TIPO"].fillna(0)
data2
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Backward fill  - poliza
data2["POLIZA"] =data2["POLIZA"].fillna(method="bfill")
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Valor en concreto - póliza
data2["POLIZA"]= data2["POLIZA"].fillna("SIN PÓLIZA")
data2
```

```python
#Comprobar
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Descargar
data2.to_csv("2020_sin_nulos.csv")
```
