```python
import pandas as pd
df = pd.read_excel('/content/Gastos y costos 20-23.xlsx', sheet_name='2021',
    skiprows=3
)
```

```python
null_summary = df.isnull().sum()
display(null_summary)
null_percentage = (df.isnull().sum() / len(df)) * 100
null_report = null_percentage[null_percentage > 0]
null_report
```

```python
df = df.drop(columns=['POLIZA'])
display(df.head())

'''
Eliminar poliza ya que tiene muchos valores nulos ~78%
'''
```

```python
df['FOLIO'] = df['FOLIO'].fillna('SIN_FOLIO').astype(str)
```

```python
df['FOLIO'] = df['FOLIO'].fillna('SIN_FOLIO')

'''
Rellenamos los valores nulos con un identificador de sin folio
'''
```

```python
#Cambiamos los nulos por
mask_mp_null = df['MP'].isna()
df.loc[mask_mp_null, 'MP'] = 'PAGOS'

display(df.isnull().sum())

'''
Cambiamos los valores nulos de MP por PAGOS cuando la descripcion de MP es pago
'''
```

```python
print(df.columns)
```

```python
import pandas as pd

# Convertirlas a numéricas (si vienen como texto)
df["RET ISR"] = pd.to_numeric(df["RET ISR"], errors="coerce").fillna(0)
df["RET IVA"] = pd.to_numeric(df["RET IVA"], errors="coerce").fillna(0)

# Crear una columna con el total
df["RET TOTAL"] = df["RET ISR"] + df["RET IVA"]

# Verificar resultado
print(df[["RET ISR", "RET IVA", "RET TOTAL"]].head())
```
